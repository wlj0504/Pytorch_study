### 2.向量concatenate（拼接）

API接口为torch.cat()，演示为两种接法，dim=0是按照形状的第一个元素对接，即为行。dim=1表示按照形状的第二个元素对接，即为列。

```python
X=torch.arange(12,dtype=torch.float32).reshape((3,4))
Y=torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
torch.cat((X,Y),dim=0),torch.cat((X,Y),dim=1)
```

同时我们可以用X==Y来验证是否在各个位置相同

```python
X==Y
```

### 2.1.3 broadcasting mechanism（广播机制）

用来解决形状不同的两个张量的操作

基本操作方法：

1.适当复制元素来进行扩充，使得两个张量具有相同的形状

2.对生成的数组执行按照元素的操作

```python
a=torch.arange(4).reshape(4,1)
b=torch.arange(2).reshape(1,2)
a,b
```

可以直接写加法，它会自动广播为4×2矩阵

```python
a+b
```

### 2.1.4索引和切片
类似于python数组，我们可以索引元素

会用“，”来去区分切片的维度

```python
# 0:3 表示切片：包含索引 0，但不包含索引 3 (即第 0, 1, 2 行)
a[-1], Y[0:3,:-2]
```


负数索引允许我们从张量的末尾向前索引：
- `-1`：最后一个元素
- `-2`：倒数第二个元素

切片公式依然遵循 `[start:end]`，即“包含开始，不包含结束”。


```python
# 创建一个包含 0 到 9 的向量
z = torch.arange(10)
print(f"原始张量: {z}")

# 选取从索引 2 到倒数第 1 个元素 (不包含倒数第 1 个)
print(f"z[2:-1]: {z[2:-1]}")

# 选取最后 3 个元素
print(f"z[-3:]: {z[-3:]}")

# 选取从开始到倒数第 2 个元素 (不包含倒数第 2 个)
print(f"z[:-2]: {z[:-2]}")

# 在矩阵中使用负索引 (选取最后两行)
print(f"Y 最后两行:\n{Y[-2:]}")
```

我们还可以通过指定索引来将元素写入矩阵。

```python
X[1,2]=9
X
```

如果想为多个元素赋值相同的数值，我们可以索引所有元素，然后赋值。

```python
X[0:2,:]=13
X
```

### 2.1.5节省内存
我们使用Y=Y+X这种的操作方式，是将X和Y的值加和，并存储在一个全新的地址中占据内存，使得Y的指针指向这个新的地址。下例为验证：

```python
before=id(Y)
Y=Y+X
id(Y)==before
```

我们可以使用切片赋值的方法，来实现节省内存


```python
before=id(Y)
Y[:]=Y+X #Y+=X
id(Y)==before
```

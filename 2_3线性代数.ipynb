## 2.3.5 张量算法的基本性质

张量的运算是在各个元素上的运算。

````python
A*B
````

````python
a=2
X=torch.arange(24).reshape(2,3,4)
a+X,a*X,(a*X).shape
````

## 2.3.6 降维

对于一维张量（向量）中的元素进行加和。


````python
x =torch.arange(4,dtype=torch.float32)
x,x.sum()
````

二维张量（矩阵）的元素和。

````python
A.shape,A.sum()
````

我们可以控制使得所有行的元素合并在不同轴，例如axis=0是列的数合并，axis=1是将每一行的数合并。

````python
A_sum_axis0=A.sum(axis=0)
A_sum_axis0,A_sum_axis0.shape
````

````python
A_sum_axis1=A.sum(axis=1)
A_sum_axis1,A_sum_axis1.shape
````

````python
A.sum(axis=[0,1]) ##沿着行和列一起求和，结果和A.sum()相同
````

求平均(mean/average)


````python
A.mean(),A.sum()/A.numel()
````

同样也可以按照指定轴降低维度

````python
A.mean(axis=0),A.sum(axis=0)/A.shape[0]
````

 axis=0是按照行进行求和，即每一列从上到下加和起来，压缩为一列。

 axis=1是按照列进行求和，保留行，即每一行从左到右加和起来，然后shape变成（5，），但是这时候会进行处理，变成shape(1,5)样子，这时候我们想保持原样就需要补充keepdims=true。

````python
import torch
A=torch.arange(20,dtype=torch.float32).reshape(5,4)
sum_A=A.sum(axis=1,keepdims=True)
sum_A
````

````python
A/sum_A
````

其实上述就是每一行的归一化操作！


一个新的函数：cumsum()函数，这个函数不会按照任意轴的维度降低输入张量的维度。他会在每一行显示前置的求和。

````python
A.cumsum(axis=0)
````

## 2.3.9 范数
范数（norm）

我们定义范数f(x),需要满足一些性质：

1.f(αx)=|α| f(x)

2.f(x+y)=f(x)+f(y)

3.f(x)>=0,f(x)==0 iff 任意x的分量都为0

一般我们默认L2范数，即为||x||

````python
u=torch.tensor([3.0,-4.0])
u.norm()
````

当然也是会经常有L1范数存在，L1范数相比较与L2范数来说收到异常值的影响比较小，计算L1函数时，我们需要如此：

````python
torch.abs(u).sum()
````

abs（）是绝对值函数，sum是求和函数，所以结合起来就是L1范数。

当然矩阵也会有范数，比较经典的是Frobenius范数，是将矩阵内的所有元素平方加和，然后进行开根号。


````python
torch.norm(torch.ones((4,9)))
````
